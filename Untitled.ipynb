```python
# Instala a biblioteca oficial da Ultralytics
!pip install ultralytics

from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow # Para exibir imagens no Colab
```

```python
# Carrega o modelo YOLO11 Nano (leve e rápido para testes)
model = YOLO('yolo11n.pt')

# Executa a detecção em uma imagem de exemplo (pode ser uma URL ou arquivo local)
# Vamos usar uma imagem de rua comum para validar as classes COCO
results = model.predict(source='https://gringo.com.vc/wp-content/uploads/2022/06/dicas-para-motociclistas.jpeg', save=True)

# Exibe o resultado na tela do Colab
for r in results:
    im_array = r.plot()  # Plota as caixas delimitadoras e labels
    cv2_imshow(im_array)
```
